```python
import seaborn as sns
```

```python
# Apply the default theme
sns.set_theme()

# Load an example dataset
df = sns.load_dataset("tips")
df2 = sns.load_dataset("iris")

df.head()     # head - returns the first 5 rows if a number is not specified.
df2.head(20)
# tip - dependent feature
# rest - independent features
```

```python
# Include only numeric features for correlation calculation
numeric_features = df.select_dtypes(include=['number']).columns
correlation_matrix = df[numeric_features].corr()

correlation_matrix

# These are numerical values
```

```python
# Heatmap - visulaize correlations b/w numeric variables
sns.heatmap(correlation_matrix)
```

# JointPlot

A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

## Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')      # kind = reg/hex
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')      # kind = reg/hex
```

# Pair plot

A “pairs plot” is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

# Dist plot

Dist plot helps us to check the distribution of the columns feature

```python
sns.histplot(df['tip'])
```

```python
sns.displot(df['tip'],kde=True,bins=10)
#  displot - This function has been deprecated and will be removed in seaborn v0.14.0. It has been replaced by histplot() and displot(
```

# Count plot

```python
# sns.countplot(data=df)
sns.countplot(data=df,x='sex')
sns.countplot(data=df,x='smoker')
sns.countplot(data=df,x='day')
```

# Bar plot

```python
sns.barplot(df, y="total_bill", x="sex")
```

# Box Plot

```python
sns.boxplot(data=df, x = 'smoker', y ='total_bill')
```

```python
sns.boxplot(data=df,orient='v')
```

# Violin Plot

Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(x=df["sex"], y=df["total_bill"])
```
